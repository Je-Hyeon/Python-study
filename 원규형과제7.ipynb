```python
# 문제 7

# 트리 구현
class TreeNode():
    def __init__(self):
        self.left = None
        self.data = None
        self.right = None

#전회
def preorder(node):
    if node == None:
        return
    print(node.data, end='->')
    preorder(node.left)
    preorder(node.right)
# 중회
def inorder(node):
    if node == None:
        return
    inorder(node.left)
    print(node.data, end='->')
    inorder(node.right)
# 후회
def postorder(node):
    if node == None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data, end='->')

# 트리 생성

node1 = TreeNode()
node1.data = 96

# 2층
node2 = TreeNode()
node2.data = 57
node1.left = node2

node3 = TreeNode()
node3.data = 153
node1.right = node3

# 3층
node4 = TreeNode()
node4.data = 12
node2.left = node4

node5 = TreeNode()
node5.data = 78
node2.right = node5

node6 = TreeNode()
node6.data = 124
node3.left = node6

node7 = TreeNode()
node7.data = 181
node3.right = node7

# 4층
node8 = TreeNode()
node8.data = 1
node4.left = node8

node9 = TreeNode()
node9.data = 49
node4.right = node9

node10 = TreeNode()
node10.data = 83
node5.right = node10

node11 = TreeNode()
node11.data = 103
node6.left = node11

node12 = TreeNode()
node12.data = 145
node6.right = node12

node13 = TreeNode()
node13.data = 178
node7.left = node13

node14 = TreeNode()
node14.data = 213
node7.right = node14
```
